```python
import pandas as pd
df = pd.read_csv("../files/output/solicitudes_de_credito.csv", sep=";", low_memory=False)
```

```python
# Filtrar el DataFrame para estrato 1 o 2
# df_filtrado = df[df["estrato"].isin([1, 2])]
df_filtrado = df

# Mostrar value_counts de cada columna
for col in df_filtrado.columns:
    print(f"Value counts para la columna '{col}':")
    print(df_filtrado[col].value_counts(dropna=False))
    print("\n" + "-"*50 + "\n")
```

```python
# Asegúrate de que la columna sea tipo fecha
# df['fecha_de_beneficio'] = pd.to_datetime(df['fecha_de_beneficio'], format='%d/%m/%Y', errors='coerce')

# Cuenta la cantidad de solicitudes por fecha
solicitudes_por_fecha = df['fecha_de_beneficio'].value_counts().sort_index()

# Estadísticas descriptivas
print(solicitudes_por_fecha.describe())
print(solicitudes_por_fecha.value_counts())

# Visualización con boxplot
import matplotlib.pyplot as plt
plt.boxplot(solicitudes_por_fecha)
plt.title('Boxplot de solicitudes por fecha de beneficio')
plt.ylabel('Cantidad de solicitudes')
plt.show()
print(df['sexo'].describe())
```
